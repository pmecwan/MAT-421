Example: Find the linear interpolation at x = 0.5.
 based on the data x = [0,1,2], y = [3,1,2]

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt


x = [0, 1, 2]
y = [3, 1, 2]

f = interp1d(x,y)
y_hat = f(0.5)
print(y_hat)
```
