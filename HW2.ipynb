Cubic Spline Interpolation



Cubic spline interpolation uses a series of piecewise cubic functions to create the interpolating curve. Each pair of adjacent points, 𝑥 𝑖 x i​and 𝑥 𝑖 + 1 x i+1​, is connected by a cubic polynomial 𝑆 𝑖 ( 𝑥 ) S i​(x), which is valid over the interval [ 𝑥 𝑖 , 𝑥 𝑖 + 1 ] [x i​,x i+1​] for 𝑖 = 1 , … , 𝑛 − 1 i=1,…,n−1.

To construct the interpolating function, we need to determine the coefficients 𝑎 a, 𝑏 b, 𝑐 c, and 𝑑 d for each cubic polynomial. Since there are 𝑛 n points, this results in 𝑛 − 1 n−1 cubic functions, each with four coefficients, leading to a total of 4 ( 𝑛 − 1 ) 4(n−1) equations

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt

x = [0, 1, 2]
y = [3, 7, 5]
f = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 2, 100)
y_new = f(x_new)

plt.figure(figsize = (10,8))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
