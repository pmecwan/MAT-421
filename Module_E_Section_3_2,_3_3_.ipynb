A limit defines the value a function approaches as the input approaches a point, the derivative measures the sensitivity of a function's output to changes in its input, and continuity ensures a function has no abrupt changes. 3.2.3 Taylor theorem

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

h = 0.1
x = np.arange(0, 2*np.pi, h)
y = np.cos(x)

forward_diff = np.diff(y)/h
x_diff = x[:-1:]
exact_solution = -np.sin(x_diff)

plt.figure(figsize = (12, 8))
plt.plot(x_diff, forward_diff, '--', label = 'Finite difference approximation')
plt.plot(x_diff, exact_solution,     label = 'Exact solution')
plt.legend()
plt.show()

max_error = max(abs(exact_solution - forward_diff))
print("Max Error: " + str(max_error))
```

3.3.3 Optimization with gradient descent

A global minimizer is a point where the function attains its absolute lowest value across its entire domain.

A local minimizer is a point where the function attains its lowest value within a small neighborhood, but this value may not necessarily be the absolute lowest value of the function globally.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

def f(x):
    return x**4 - 4*x**2 + x

x_values = np.linspace(-2, 2, 500)
y_values = f(x_values)

plt.plot(x_values, y_values, label='f(x) = x^4 - 4x^2 + x')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of the function')
plt.grid(True)
plt.legend()

result_global = minimize(f, x0=0)
x_global_min = result_global.x
y_global_min = result_global.fun

plt.plot(x_global_min, y_global_min, 'ro', label=f'Global Minimizer: ({x_global_min[0]:.2f}, {y_global_min:.2f})')

result_local = minimize(f, x0=2)
x_local_min = result_local.x
y_local_min = result_local.fun

plt.plot(x_local_min, y_local_min, 'go', label=f'Local Minimizer: ({x_local_min[0]:.2f}, {y_local_min:.2f})')
plt.legend()
plt.show()
```
